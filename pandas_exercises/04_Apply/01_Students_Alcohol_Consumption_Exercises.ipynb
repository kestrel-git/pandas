### Step 9. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
def is_majority(age: int) -> bool:
  '''
  주어진 나이가 성년인지 판단한다.

  Args:
    age (int): 판별할 나이

  Returns:
    bool: 나이가 17세보다 많으면 True, 그렇지 않으면 False
  '''
  return age > 17
```

```python
df['legal_drinker'] = df['age'].apply(is_majority)
df.head()
```
